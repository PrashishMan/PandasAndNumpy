<h3> Filtering with string methods </h3>

```python
bond[bond["Actor"].str.lower() == 'sean connery']
```

```python
mask = bond['Actor'].str.lower().str.contains("sean")
bond[mask]
```

```python
mask2 = bond["Actor"].str.lower().str.startswith("d")
bond[mask2]
```

```python
mask3 = bond["Director"].str.lower().str.endswith('g')
bond[mask3]
```

```python
b = bond["Actor"].str.split(" ", expand = True)
b.columns = ["First Name", "Last Name"]
```

```python
bond[b.columns] = b
```

```python
b3 = bond.drop("Actor",axis = 1).copy()
```
