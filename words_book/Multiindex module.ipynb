# The pivot_table() method

```python
foods = pd.read_csv("./../data/foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "City"], aggfunc="sum")
```

```python
foods.pivot_table(values = "Spend", index = ["City", "Item"],columns = ["Frequency","Gender"], aggfunc="sum")
```

```python
foods.pivot_table(values = "Spend", index = ["City", "Item"],columns ="Gender", aggfunc="max")
```

```python
foods.pivot_table(values = "Spend", index = ["City", "Item"],columns ="Gender", aggfunc="count")
```
