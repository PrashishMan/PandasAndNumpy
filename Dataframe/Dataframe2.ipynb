<h3> Filtering with a where method </h3>

```python
bond_3 = pd.read_csv("./../data/jamesbond.csv", index_col = "Film")
bond_3.sort_index(inplace = True)
```

```python
bond_3.head(3)
```

```python
mask = bond_3["Actor"] == "Sean Connery"
```

```python
bond_3.where(mask)
```

```python
mask_2 = bond_3["Box Office"] > 800
bond_3.where(mask & mask_2)
```

<h3>The .query() method</h3>

```python
# column names must not have spaces
bond_3.columns = [col.replace(" ", "_") for col in bond_3.columns]
```

```python
bond_3.head(2)
```

```python
bond_3.query('Actor == "Sean Connery"')
```

<b> Apply method on single columns </b>

```python
def add_millions(number):
    return str(number) + "Millions"

bond_3.loc[:,'Box_Office'] = bond_3.loc[:,'Box_Office'].apply(add_millions)
```

```python
bond_3.head()
```

```python
bond_3.loc[:, "Budget"]=bond_3.loc[:, "Budget"].apply(add_millions)
```

```python
bond_3.head()
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond_3[col] = bond_3[col].apply(add_millions) 
```

```python
bond_3.head(3)
```

<h3> Apply method with row values </h3>

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Bronson":
        return "Best"
    
    elif actor == "Roger Moore" and budget > 40:
        return "Entertaining"
    
    else:
        return "No Idea"
```

```python
bond_3["Review"] = bond_3.apply(good_movie, axis = "columns")
```

```python
bond_3.head()
```
